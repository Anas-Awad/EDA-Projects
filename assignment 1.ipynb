### Use the following lists to build the data frame below using dictionary method as we done in the session

```python
data = {
    'students' :[ 'Naomi', 'Yumi','Hinata','Hana', 'Akira', 'Yuki','Yudi', 'Sakura', 'Sasuke','Hiroshi', 'Ken', 'Emi'],
    'mathematics' : [1, 3, 1, 3, 4, 3, 1, 9, 5, 9, 1, 2],
'chemestry' : [6, 5, 5, 8, 6, 6, 5, 3, 7, 8, 6, 4],
'biology' : [8, 7, 9, 9, 7, 7, 8, 7, 7, 7, 7, 9],
'economy' : [3, 2, 7, 8, 9, 9, 5, 9, 8, 6, 7, 9]}
```

```python
df = pd.DataFrame(data)
```

# create a new column to store the students' final grade

```python
 df['Final_score'] = df['mathematics'] + df['chemestry'] + df['biology'] + df['economy']
```

```python
df
```

## create a new column to inform whether the student has been approved or not, based on the final grade (the grade must be greater than or equal to 25); to do so, create a function and use the pandas apply () method.

```python
df['Status'] = df['Final_score'].apply(lambda x: 'approved' if x >= 25 else 'Rejected')
```
