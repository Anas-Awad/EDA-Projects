```python
import requests
import json
import csv
```

```python
reponse = requests.get('https://coronavirus.m.pipedream.net/')
```

```python
data = json.loads(reponse.text)
```

```python
data.keys()
```

```python
columns = ['Country_Region','Confirmed','Deaths']
```

```python
with open('covidStatistics.csv','w',newline='') as f:
    writer = csv.DictWriter(f,columns)
    writer.writeheader()
    for i in data['rawData']:
        country = i['Country_Region']
        confirmed = i['Confirmed']
        deaths = i['Deaths']
        writer.writerow({columns[0]:country,columns[1]: confirmed, columns[2]:deaths})
```
